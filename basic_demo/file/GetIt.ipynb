```python
import requests
from selenium import webdriver
from selenium.webdriver.edge.service import Service
import json
import re
import csv
```

```python
r = requests.get('http://localhost:3000/playlist/catlist')       # 最基本的不带参数的get请求
cat = r.json()['sub']
cats = [c['name'] for c in cat]
```

```python
service = Service(r"D:\Hs-black\Tsinghua\NLP\project\Music\chromedriver.exe")
options = webdriver.EdgeOptions()
options.add_argument('--start-maximized')
options.service = service
```

```python
def info(id, c):
    r = requests.get('http://localhost:3000/playlist/detail?id=' + str(id))
    try:
        detail = r.json()['playlist']
        io = [id, detail['name'], detail['description'], detail['tags'] + [c]]
    except:
        print (id)
        return None
    playlist.append(id)
    r = requests.get('http://localhost:3000/playlist/track/all?id=' + str(id))
    try:
        songs = r.json()['songs']
        io.append([song['id'] for song in songs])
    except:
        print (id)
        io.append([])
    playlist.append(id)
    return io
```

```python
playlist = []
for c in cats[cats.index('爵士'):]:
    print (f"Running on Cat {c}")
    try:
        driver = webdriver.Edge(options=options)
        driver.get("https://music.163.com/#/discover/playlist/?order=hot&cat=" + c)
        driver.switch_to.frame('g_iframe')
        source = driver.page_source
        pattern = r'(\d+)</a>'
        pages = 0
        for m in re.finditer(pattern, source):
            pages = max(pages, int(m.group(1)))
        print (f"there are total {pages} pages")
        pages = min(pages, 50)
        for i in range(pages):
            try:
                with open(c  + '_'+ str(i) + ".csv", 'w', newline='', encoding='utf-8') as f:
                    print (f"Running on Cat {c} Page {i}")
                    writer = csv.writer(f)
                    driver.get("https://music.163.com/#/discover/playlist/?order=hot&cat=" + c + "&lim=35&offset=" + str(35 * i))
                    driver.switch_to.frame('g_iframe')
                    source = driver.page_source
                    pattern = r'/playlist\?id=(\d+)'
                    for m in re.finditer(pattern, source):
                        if m.group(1) in playlist:
                            continue
                        content = info(m.group(1), c)
                        if content is not None:
                            writer.writerow(content)
            except:
                print (f"Error Running on Cat {c} Page {i}")
                continue
    except:
        print (f"Error Running on Cat {c}")
        continue
```
